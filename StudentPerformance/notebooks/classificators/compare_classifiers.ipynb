```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
import glob
```

```python
results_files = sorted(glob.glob("results/*_results.csv"))
frames = []
skipped = []
for f in results_files:
    df = pd.read_csv(f)
    if "Test_F1_Macro" in df.columns:
        frames.append(df)
    else:
        skipped.append(os.path.basename(f))

if not frames:
    print("No classifier result files with 'Test_F1_Macro' found. Skipping comparison.")
else:
    comparison = pd.concat(frames, ignore_index=True)
    comparison = comparison.sort_values("Test_F1_Macro", ascending=False)
    print("Classifier Comparison Results (sorted by Test_F1_Macro):")
    print(comparison.to_string(index=False))
    os.makedirs("results", exist_ok=True)
    comparison.to_csv("results/all_classifiers_comparison.csv", index=False)
    print("\nSaved: results/all_classifiers_comparison.csv")

if skipped:
    print("Skipped files (different format):", skipped)
```

```python
metrics_cols = ["Test_Accuracy", "Test_Precision_Macro", "Test_Recall_Macro", "Test_F1_Macro"]
if 'comparison' in globals() and not comparison.empty:
    comparison[metrics_cols].plot(kind="bar", figsize=(10, 5))
    plt.title("Krahasimi i Klasifikuesve")
    plt.ylabel("Score")
    plt.xticks(range(len(comparison)), comparison["Model"], rotation=15)
    plt.legend(loc="lower right")
    plt.tight_layout()
    os.makedirs("visualizations", exist_ok=True)
    plt.savefig("visualizations/classifiers_comparison.png", dpi=300, bbox_inches="tight")
    plt.show()
else:
    print("No comparison data available to plot.")
```
